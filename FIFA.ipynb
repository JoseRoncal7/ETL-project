# FIFA ETL PROJECT

## Extract and Transform

I got two datasets from kaggle. The first one was a table of different stats of every active soccer player, and the second one was a table that shows the ranking and information about every country that is registered in FIFA.

After cleaning both tables, I created a database in PgAdmin and imported both csv tables into the tables in PgAdmin. 

### FIFA Table

```python
fifa_table = "ETL/fifa_ranking.csv"
fifa_table_df = pd.read_csv(fifa_table)
fifa_table_df.head()
```

I decided to reduce the table to Rank, Country name, Country abbreviation, and Confederation, as well as rename the columns.

```python
fifa_table_reduced = fifa_table_df[["rank","country_full","country_abrv","confederation"]]
fifa_ranking_df = fifa_table_reduced.rename(columns={"rank":"Country_Rank", "country_full":"Country","country_abrv":"Country_abbreviation","confederation":"Confederation"}) 
fifa_ranking_df.head()
```

```python
#Export as csv
fifa_ranking_df.to_csv("Ranking_fifa_table.csv",index=False)
```

### FIFA Players Table

```python
fifa_players = "ETL/FIFA19.csv"
```

```python
fifa_players_df = pd.read_csv(fifa_players)

fifa_players_df.head()
```

I decided to reduce the players table to ID, Name, Age, country, Overall Score, and Club name.

```python
fifa_reduced = fifa_players_df[["ID","Name","Age","Nationality","Overall","Club"]]

fifa_reduced.head()
```

```python
# Export as csv
fifa_reduced.to_csv("FIFA_PLAYERS2019.csv", index=False)
```
